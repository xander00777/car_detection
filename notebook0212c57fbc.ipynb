```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
```

```python
model=YOLO("yolov10x.pt")
```

```python
def predict(model, img, classes=[], conf=0.5):
    if classes:
        results = model.predict(img, classes=classes, conf=conf)
    else:
        results = model.predict(img, conf=conf)

    return results

def predict_and_detect(model, img, classes=[], conf=0.5, rectangle_thickness=2, text_thickness=1):
    results = predict(model, img, classes, conf=conf)
    for result in results:
        for box in result.boxes:
            cv2.rectangle(img, (int(box.xyxy[0][0]), int(box.xyxy[0][1])),
                          (int(box.xyxy[0][2]), int(box.xyxy[0][3])), (255, 0, 0), rectangle_thickness)
            cv2.putText(img, f"{result.names[int(box.cls[0])]}",
                        (int(box.xyxy[0][0]), int(box.xyxy[0][1]) - 10),
                        cv2.FONT_HERSHEY_PLAIN, 1, (255, 0, 0), text_thickness)
    return img, results
```

```python
from os import listdir
from os.path import isfile, join
import random 
onlyfiles = [f for f in listdir("../input/car-object-detection/data/testing_images") if isfile(join("../input/car-object-detection/data/testing_images", f))]
```

```python
 def display_faces(files):
    images=[]
    result_img=[]
    for i in range(10):
        images.append( cv2.imread(f"../input/car-object-detection/data/testing_images/{files[random.randint(1,len(files)-1)]}"))
        res, _ = predict_and_detect(model, images[i], classes=[], conf=0.5)
        result_img.append(res)

    n_cols = 2
    n_rows = 10//n_cols+1
    plt.figure(figsize=(32,80))
    
    for i in range(len(images)):
    #     print(image.shape)
    #    disp_img = np.reshape(image, (64,64,3))
        plt.subplot(n_rows, n_cols, i+1)
        plt.imshow(result_img[i])

        
```

```python
display_faces(onlyfiles)
```
